# Logistic Regression 

$ z = B0 + B1*x1 + B2*x2 + ... + Bn*xn $

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

### Step 1 - Data Ingestion (Read the csv file)

```python
import pandas as pd
df = pd.read_csv("iris.csv")
df.head()
```

### Step 2 - Perform basic data quality checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

### Step 3 - Seperate X and Y(species)

```python
X = df.drop(columns=["species"])
Y = df[["species"]]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

### Step 4 - apply preprocessing on X

```python
X.dtypes
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()
).set_output(transform="pandas")
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

```python
num_pipe
```

### Step 5 - Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.33, random_state=21)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Step 6 - Build Model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain, ytrain)
```

```python
model.classes_
```

```python
model.intercept_
```

```python
model.coef_
```

```python
xtrain.columns
```

### Step 7 - Model evaluation

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(model, xtest, ytest)
```

```python
ypred_test = model.predict(xtest)
ypred_test[0:5]
```

```python
ytest.head()
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, xtrain, ytrain, cv=5, scoring="f1_macro")
scores
```

```python
scores.mean()
```

### Clearly above model is good because f1_macro > 0.80 in test and in cross validation

### Step 7 - Out of sample prediction

```python
xnew = pd.read_csv("iris_sample.csv")
xnew
```

```python
num_pipe
```

```python
xnew_pre = num_pipe.transform(xnew)
xnew_pre
```

```python
preds = model.predict(xnew_pre)
preds
```

```python
probs = model.predict_proba(xnew_pre)
probs
```

```python
model.classes_
```

```python
xnew["species_pred"] = preds
xnew[model.classes_] = probs
xnew = xnew.round(4)
xnew
```

```python
xnew["species_pred"].value_counts()
```

```python
xnew.to_csv("Result.csv", index=False)
```
