### Step 1 - Data Ingestion

```python
import pandas as pd
df = pd.read_csv("Cars93.csv", na_values=["", "NA"], keep_default_na=False)
df.head()
```

### Step 2 - Basic data quality checks

```python
df.info()
```

```python
m = df.isna().sum()
m[m>0]
```

## Step 3 - Seperate X and Y(Weight)

```python
X = df.drop(columns=["id", "Weight"])
Y = df[["Weight"]]
```

```python
X.head()
```

```python
Y.head()
```

### Perform preprocessing on X

```python
X.dtypes
```

```python
cat = list(X.columns[X.dtypes == "object"])
con = list(X.columns[X.dtypes != "object"])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()
)
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore", sparse_output=False)
)
```

```python
pre = ColumnTransformer(
    [
        ("num", num_pipe, con),
        ("cat", cat_pipe, cat)
    ]
).set_output(transform="pandas")
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

### Step 5 - Train Test Split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=10)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

### Step 6 - Model Building and tuning

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(xtrain, ytrain)
```

```python
model1.score(xtrain, ytrain)
```

```python
model1.score(xtest, ytest)
```

### Ridge model

```python
from sklearn.linear_model import Ridge
model2 = Ridge(alpha=1)
model2.fit(xtrain, ytrain)
```

```python
model2.score(xtrain, ytrain)
```

```python
model2.score(xtest, ytest)
```

### Tune the alpha value with gridsearch

```python
import numpy as np
```

```python
params = {
    "alpha": np.arange(start=0.1, stop=100, step=0.1)
}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
model_rr = Ridge()
gscv_ridge = GridSearchCV(model_rr, param_grid=params, cv=5, scoring="r2", n_jobs=-1)
gscv_ridge.fit(xtrain, ytrain)
```

```python
gscv_ridge.best_params_
```

```python
gscv_ridge.best_score_
```

```python
best_ridge = gscv_ridge.best_estimator_
best_ridge
```

### Ridge Model Evaluation

```python
best_ridge.score(xtrain, ytrain)
```

```python
best_ridge.score(xtest, ytest)
```

```python
from sklearn.metrics import (
    root_mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score
)
```

```python
def evaluate_model(model, x, y):
    ypred = model.predict(x)
    rmse = root_mean_squared_error(y, ypred)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE : {mae:.2f}")
    print(f"MAPE : {mape:.4f}")
    print(f"R2 : {r2:.4f}")
```

```python
print("Ridge Training Results : ")
evaluate_model(best_ridge, xtrain, ytrain)
```

```python
print("Ridge Testing Results : ")
evaluate_model(best_ridge, xtest, ytest)
```

### Lasso (L1 Penalty)

```python
from sklearn.linear_model import Lasso
model3 = Lasso(alpha=1)
model3.fit(xtrain, ytrain)
```

```python
model3.score(xtrain, ytrain)
```

```python
model3.score(xtest, ytest)
```

### Tune alpha value

```python
params2 = {
    "alpha": np.arange(start=0.1, stop=100, step=0.1)
}
params2
```

```python
model_lr = Lasso()
gscv_lasso = GridSearchCV(model_lr, param_grid=params2, cv=5, scoring="r2", n_jobs=-1)
gscv_lasso.fit(xtrain, ytrain)
```

```python
gscv_lasso.best_params_
```

```python
gscv_lasso.best_score_
```

```python
best_lasso = gscv_lasso.best_estimator_
best_lasso
```

### Evaluate Lasso

```python
best_lasso.score(xtrain, ytrain)
```

```python
best_lasso.score(xtest, ytest)
```

```python
print("Lasso Train Results :")
evaluate_model(best_lasso, xtrain, ytrain)
```

```python
print("Lasso Test Results :")
evaluate_model(best_lasso, xtest, ytest)
```
